## Now Lets Look at R

Now why do we even bother to use R?

Well for statistical models and simulations it is simply better and easier to use, and plotting the results for that is also essential.

Now we will focus on two main tools in R to plot, one is Base-R and one is ggplot2

Base-R is inbuilt into R. It provides a set of fundamental tools for data handling and plotting directly accessible without additional packages and it is easier to learn.

ggplot2 is more the fancy schmancy stuff, developed by Hadley Wickham. It follows the grammar of graphics paradigm, providing more flexible plots.

now lets take a look at ggplot2

Using examples from [R-Stats](http://r-statistics.co/Top50-Ggplot2-Visualizations-MasterList-R-Code.html)



```r
library(ggplot2)
```

```r
data(mpg, package="ggplot2")

theme_set(theme_bw())
g <- ggplot(mpg, aes(cty, hwy))
g + geom_jitter(width = .5, size=1) +
  labs(subtitle="mpg: City vs Highway mileage",
       y="Highway",
       x="City",
       title="Scatter Plot Example")
```

```r
library(ggplot2)
data(mpg, package="ggplot2")

mpg_select <- mpg[mpg$manufacturer %in% c("audi", "ford", "honda", "hyundai"), ]

theme_set(theme_bw())
g <- ggplot(mpg_select, aes(displ, cty)) +
  labs(subtitle="mpg: Displacement vs City Mileage",
       title="Bubble chart")

g + geom_jitter(aes(col=manufacturer)) +
  geom_smooth(aes(col=manufacturer), method="lm", se=F)
```

```r
library(lubridate)
theme_set(theme_bw())

df <- economics[, c("date", "psavert", "uempmed")]
df <- df[lubridate::year(df$date) %in% c(1967:1981), ]

brks <- df$date[seq(1, length(df$date), 12)]
lbls <- lubridate::year(brks)

ggplot(df, aes(x=date)) +
  geom_area(aes(y=psavert+uempmed, fill="psavert")) +
  geom_area(aes(y=uempmed, fill="uempmed")) +
  labs(title="Area Chart of Returns Percentage",
       subtitle="From Wide Data format",
       caption="Source: Economics",
       y="Returns %") +
  scale_x_date(labels = lbls, breaks = brks) +
  scale_fill_manual(name="",
                    values = c("psavert"="#00ba38", "uempmed"="#f8766d")) +
  theme(panel.grid.minor = element_blank())
```
